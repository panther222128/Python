## Boolean Selection, 조건에 맞는 데이터 선택

### class_ = train_data['Pclass'] == 1
### age_ = (train_data['Age'] >= 30) & (train_data['Age'] < 40)

### 위 두 가지 Boolean에 해당하는 데이터만 추출하는 경우

### train_data[class_ & Age_]

```python
class_ = train_data['Pclass'] == 1
class_
```

```python
age_ = (train_data['Age'] >= 30) & (train_data['Age'] < 40)
age_
```

```python
train_data[class_ & age_]
```

## 앞서 본 것이 Boolean 반환 두 가지를 조건으로 만들어 데이터를 반환하는 것이라면
## 그것과 관계 없이 조건에 맞는 데이터를 바로 반환할 때는
## train_data[(train_data['Age'] >= 30) & (train_data['Age'] <40)]
## train_data[train_data['Age'] < 26]

```python
train_data[(train_data['Age'] >= 30) & (train_data['Age'] < 40)]
```

```python
train_data[train_data['Age'] < 26]
```
